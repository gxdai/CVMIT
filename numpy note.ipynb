```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.legend(["Sin", "Cos"])
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.title("SIN")
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title("COS")
plt.show()
```

```python
from scipy.signal import convolve2d as conv2d
import cv2
```

```python
image = cv2.imread("city.jpeg")
```

```python
v_edge = np.array([-1, 0, 1]).reshape(1, 3)
h_edge = v_edge.reshape(3, 1)
k_size = 15
blur = np.ones((k_size, k_size)) / (k_size ** 2)

sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

edge = np.array([
    [0, -1, 0],
    [-1, 4, -1],
    [0, -1, 0]
])
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).astype(np.float32) / 255.0

def conver_imagge(image):
    return (np.clip(image, 0, 1.0) * 255.0).astype(np.uint8)

v_edge_image = conver_imagge(conv2d(gray_image, v_edge))
h_edge_image = conver_imagge(conv2d(gray_image, h_edge))
blur_image = conver_imagge(conv2d(gray_image, blur))
sharp_image = conver_imagge(conv2d(gray_image, sharpen))
edgeimage = conver_imagge(conv2d(gray_image, edge))
```
